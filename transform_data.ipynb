```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
csv_file_path = "products_data.csv"
df = pd.read_csv(csv_file_path)
df
```

```python
json_column_name = 'prices'  # Replace 'your_column_name' with the actual column name
 
# Function to extract prices for 'Market Price' and 'Lenskart Price'
def extract_prices(json_str):
    data_list = json.loads(json_str.replace("'", "\""))
    market_price = next((item['price'] for item in data_list if item['name'] == 'Market Price'), None)
    lenskart_price = next((item['price'] for item in data_list if item['name'] == 'Lenskart Price'), None)
    return pd.Series({'Market_Price': market_price, 'Lenskart_Price': lenskart_price})
 
# Apply the function to the specified column and create new columns with the extracted prices
df[['Market_Price', 'Lenskart_Price']] = df[json_column_name].apply(extract_prices)
df = df.drop('prices',axis=1)
# Print the updated DataFrame
print(df)
 
# Save the updated DataFrame to a new CSV file
df.to_csv('output_file.csv', index=False)
```

```python
# Specify the column containing the JSON data
json_column_name = 'hashtagList'  # Replace 'hashtagList' with the actual column name
 
# Function to extract 'name' values and join them into a string
def extract_names(json_str):
    data_list = json.loads(json_str.replace("'", "\""))
    name_values = [item.get('name', '') for item in data_list]
    return ",".join(name_values)
 
# Apply the function to the specified column and create a new column with the converted values
df[json_column_name] = df[json_column_name].apply(extract_names)
 
# Print the updated DataFrame
print(df)
 
# Save the updated DataFrame to a new CSV file
df.to_csv('output_file.csv', index=False)
 
```
